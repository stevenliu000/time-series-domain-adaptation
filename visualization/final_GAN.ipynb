```python
import os
import numpy as np
import matplotlib.pyplot as plt

path = "/home/tianqinl/Code/time-series-domain-adaptation/train_related/test/Linear_GAN/FINAL_GAN/lbl_percent_0.700000/task_3E_gpweight_5.000000_ncritic_0.080000_sclass_0.700000_global_1_dglobal_0.010000_local_0_dlocal_0.010000"

source_test = np.load(path + "/source_acc_unlabel_.npy")
target_test = np.load(path + "/target_acc_unlabel_.npy")
epochs = [i for i in range(len(source_test)) if i %2==0]

plt.plot(epochs, target_test[epochs])
plt.plot(epochs, source_test[epochs])
plt.xlim(-1,2000)
plt.legend(["target", "source"])
plt.title("Global GAN Acc")
```
