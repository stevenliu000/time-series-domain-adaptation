```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
model_path = "/home/tianqinl/Code/time-series-domain-adaptation/train_related/Linear_GAN/SimpleMLP_task_3E_gpweight_100.000000_dlocal_0.010000_critic_0.160000_rand_0_sclass_0.700000_global_0_local_1"
source_acc_label_ = np.load(os.path.join(model_path, "source_acc_label_.npy"))
source_acc_unlabel_ = np.load(os.path.join(model_path, "source_acc_unlabel_.npy"))
target_acc_label_ = np.load(os.path.join(model_path, "target_acc_label_.npy"))
target_acc_unlabel_ = np.load(os.path.join(model_path, "target_acc_unlabel_.npy"))
```
