```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python

with open('./data_unzip/processed_file_3Av2.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
data.keys()
```

```python
tr_data = data['tr_data']
tr_lbl = data['tr_lbl']
te_data = data['te_data']
te_lbl = data['te_lbl']
```

```python
print("Tr_tr shape", tr_data.shape)
print("Tr_label", tr_lbl.shape)
print("te_data", te_data.shape)
print("te_lbl", te_lbl.shape)
```

```python
# mean over all sample for different time xs

mean_tr_data = np.mean(tr_data, axis=0)
## histogram
plt.subplot(121)
plt.xlim(-0.06, 0.20)
plt.hist(mean_tr_data[:,0], bins=100)
plt.subplot(122)
plt.xlim(-0.06, 0.20)
plt.hist(mean_tr_data[:,1], bins=100)
plt.show()
```

```python
## mean over all sample for different time xs values
plt.subplot(121)
plt.ylim(-0.10, 0.20)
plt.scatter([i for i in range(1600)], mean_tr_data[:,0])
plt.subplot(122)
plt.ylim(-0.10, 0.20)
plt.scatter([i for i in range(1600)], mean_tr_data[:,1])
```

```python
## values finer resolution
plt.subplot(121)
plt.ylim(-0.10, 0.20)
plt.scatter([i for i in range(25)], mean_tr_data[0:25,0])
plt.subplot(122)
plt.ylim(-0.10, 0.20)
plt.scatter([i for i in range(25)], mean_tr_data[0:25,1])
```

```python
mean_tr_data[1:100,0].shape
```

```python
# examine each indivisual 

mean_inv_tr_data = np.mean(tr_data, axis=1)
```

```python
## histogram for different data point timeseries value
plt.subplot(121)
plt.hist(mean_inv_tr_data[:,0], bins=100)
plt.subplot(122)
plt.hist(mean_inv_tr_data[:,1], bins=100)
plt.show()
```

```python
## x values
plt.subplot(121)
plt.ylim(-0.15, 0.20)
plt.scatter([i for i in range(5000)], mean_inv_tr_data[:,0])
plt.subplot(122)
plt.ylim(-0.15, 0.20)
plt.scatter([i for i in range(5000)], mean_inv_tr_data[:,1])
```
