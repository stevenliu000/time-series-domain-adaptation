```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
plt.rcParams['figure.figsize'] = [7, 5]
```

```python
naive_adaption_path = "/home/weixinli/time-series-domain-adaptation/train_related/f-gan/naive_adaption_KL/"
unlabeled_f_div = np.load(naive_adaption_path+"unlabeled_KL.npy")
labeled_f_div = np.load(naive_adaption_path+"labeled_KL.npy")
epochs = np.load(naive_adaption_path+"epochs.npy")
```

```python
naive_adaption_classifier_path = "/home/weixinli/time-series-domain-adaptation/train_related/f-gan/f-gan-old/naive_adaptation_JS_classifier/"
unlabeled_target_acc = np.load(naive_adaption_classifier_path+'acc_target_unlabeled_classifier_.npy')
unlabeled_source_acc = np.load(naive_adaption_classifier_path+'acc_source_unlabeled_classifier_.npy')
epochs_acc = np.load(naive_adaption_classifier_path+'epochs.npy')
```

```python
plt.plot(epochs_acc, unlabeled_target_acc, label='target')
plt.plot(epochs_acc, unlabeled_source_acc, label='source')
plt.title('Accuracy of unlabeled data, naive adapatation', fontsize=13)
plt.legend(loc='lower right', fontsize=14)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylim([0.1,0.5])
plt.ylabel("accuracy", fontsize=13)
plt.xlabel('epoch', fontsize=13)
```

```python
plt.plot(epochs, unlabeled_f_div)
plt.title('KL-div of unlabeled data, naive adapatation', fontsize=13)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel("KL-div", fontsize=13)
plt.xlabel('epoch', fontsize=13)
```

```python
plt.plot(epochs, labeled_f_div)
plt.title('KL-div of labeled data, naive adapatation', fontsize=13)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel("KL-div", fontsize=13)
plt.xlabel('epoch', fontsize=13)
```
